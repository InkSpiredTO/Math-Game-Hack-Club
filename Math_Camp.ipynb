```python
import random
Play = True
CamEnergy = 0
Location = "Campgrounds"
MyPhotos = {
  "Blue jay": 0,
  "Bunny": 0,
  "Canada Lynx": 0,
  "Deer": 0,
  "Duck": 0,
  "Eagle": 0,
  "Fox": 0,
  "Moose": 0,
  "Raven": 0,
  "Robin": 0,
  "Salmon": 0,
}
Choice = ""
Campgrounds = []
Forest = [""]
Lake = []
Mountains = []

print("Welcome to...\n")
print("ðŸŒ² MATH NATIONAL PARK ðŸŒ²:\n")
print("Photograph animals and get know this amazing place!")
input("Press enter to start your camping journey: ")

print("\nYou start your journey at the campgrounds.\nWhat do you do?\n")

def Travel():
  print("\nWhere do you want to travel:\n 1) Campgrounds\n 2) Forest \n 3) Lake\n 4) Mountains")
  global Choice
  Choice = input(":")
  if Choice == "1":
    Location = "Campgrounds"
  elif Choice == "2":
    Location = "Forest"
  elif Choice == "3":
    Location = "Lake"
  elif Choice == "4":
    Location = "Mountains"

def Photos():
  if CamEnergy != 0:
    if Location == "Campgrounds":
      print("You are in the campgrounds no animals here.")
    elif Location == "Forest":
      print()
    elif Location == "Lake":
      print()
    elif Location == "Mountains":
      print()
  else:
    print("You have no energy. Answer math questions to recharge:")
    Energy()

def Album():
  print("Here is your photo collection:\n",MyPhotos)

def Energy():
  print()
  Var1 = random.randint(1,10)
  Var2 = random.randint(1,10)
  ans = input(f"What is {Var1} x {Var2}? \n:")
  if ans == str(Var1 * Var2):
    print("Correct!")
    global Energy
    CamEnergy += 1

while Play:
  print(f"Statuses -- | Camera Battery: {CamEnergy} | ")
  print("Do you want to:\n 1) Travel\n 2) Photograph animals\n 3) Shop at nearest trading post\n 4) My Photos\n 5) My stuff")
  Choice = input(":")
  if Choice == "1":
    print()
    Travel()
    print("\n")
  elif Choice == "2":
    print()
    Photos()
    print("\n")
  elif Choice == "3":
    print()
    Album()
    print("\n")
```

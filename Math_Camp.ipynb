```python
# import libraries
import random
import time

# define AMAZING VARIABLES (as well as dictionaries and lists)
Play = True
CamEnergy = 0
Location = "Forest"
MyPhotos = {
  "Blue jay": 0,
  "Bunny": 0,
  "Canada Lynx": 0,
  "Deer": 0,
  "Duck": 0,
  "Eagle": 0,
  "Fox": 0,
  "Moose": 0,
  "Raven": 0,
  "Robin": 0,
  "Salmon": 0,
  "Mountain Goat": 0,
}

# Lists containing animals
Forest = []
for i in range(5):
    Forest.append("Moose")
for i in range(10):
    Forest.append("Blue jay")
for i in range(20):
    Forest.append("Bunny")
for i in range(7):
    Forest.append("Deer")
for i in range(10):
    Forest.append("Eagle")
for i in range(10):
    Forest.append("Fox")
for i in range(10):
    Forest.append("Raven")
for i in range(10):
    Forest.append("Robin")

Lake = []
for i in range(20):
    Lake.append("Duck")
for i in range(20):
    Lake.append("Salmon")
for i in range(5):
    Lake.append("Moose")

Mountains = []
for i in range(5):
    Forest.append("Canada lynx")
for i in range(10):
    Forest.append("Eagle")
for i in range(20):
    Forest.append("Robin")
for i in range(20):
    Forest.append("Raven")
for i in range(5):
    Forest.append("Mountain Goat")

# Intro
print("Welcome to...\n")
print("ðŸŒ² MATH NATIONAL PARK ðŸŒ²:\n")
print("Photograph animals and get know this amazing place!")
input("Press enter to start your camping journey: ")

print("\nYou start your journey at the campgrounds.\nWhat do you do?\n")

# AMAZING Functions
def Travel():
  print("Where do you want to travel:\n 1) Forest \n 2) Lake\n 3) Mountains")
  Choice = input(":")
  global Location
  if Choice == "1":
    Location = "Forest"
  elif Choice == "2":
    Location = "Lake"
  elif Choice == "3":
    Location = "Mountains"

  print(f"You are now in the {Location}.")

def Photos():
  if CamEnergy != 0:
    if Location == "Forest":
      PhotoMode()
    elif Location == "Lake":
      PhotoMode()
    elif Location == "Mountains":
      PhotoMode()
  else:
    print("You have no energy.")
    Energy()

def Album():
  print("Here is your photo collection:\n",MyPhotos)

def Energy():
  print()
  print("Answer the following question to recharge:")
  Var1 = random.randint(1,10)
  Var2 = random.randint(1,10)
  ans = input(f"What is {Var1} x {Var2}? \n:")
  if ans == str(Var1 * Var2):
    print("Correct!")
    global CamEnergy
    CamEnergy += 1

def PhotoMode():
  global CamEnergy
  if Location == "Forest":
    Animal = random.choice(Forest)
  elif Location == "Lake":
    Animal = random.choice(Lake)
  elif Location == "Mountains":
    Animal = random.choice(Mountains)

  MyPhotos[Animal] += 1
  CamEnergy -= 1
  print(f"You have taken a photo of a {Animal}")

# This while loop repeats FOREVER until user quits the game
while Play:
  print(f"Statuses -- | Camera Battery: {CamEnergy} | Location: {Location}")
  print("Do you want to:\n 1) Travel\n 2) Photograph animals\n 3) Photo Collection\n 4) Power battery\n x - Quit game")
  Choice = input(":")
  if Choice == "1":
    print()
    Travel()
    print("\n")
  elif Choice == "2":
    print()
    Photos()
    print("\n")
  elif Choice == "3":
    print()
    Album()
    print("\n")
  elif Choice == "4":
    print()
    Energy()
    print("\n")
  elif Choice == "x":
    Play = False
```
